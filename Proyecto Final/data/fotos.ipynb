```python
import pandas as pd
import requests
```

```python
df = pd.read_csv('restaurants.csv')
```

```python
df.head()
```

```python
def search_picture(term, downloaded_images):
    url = "https://api.pexels.com/v1/search"
    api_key = ""
    headers = { "Authorization": api_key }
    params = { "query": term, "per_page": 1 }
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        response = response.json()
        if response['total_results'] > 0:
            url = response['photos'][0]['src']['original']
            filename = url.split('/')[-1]
            if filename in downloaded_images:
                return filename
            response = requests.get(url)
            open(f"images/{filename}", "wb").write(response.content)
            downloaded_images.append(filename)
            return filename
```

```python
donwloaded_images = []
df = df.assign(Image=df['Name'].apply(lambda x: search_picture(x, donwloaded_images)))
df.head()
```

```python
df.to_csv("restaurants.csv", index=False)
```
